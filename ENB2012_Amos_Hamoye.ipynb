```python
import numpy as np
import pandas as pd
```

```python
# import csv file from html:
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
energydata = pd.read_csv(url)
energydata.head()
energydata.to_csv('energydata_complete.csv')
```

```python
# dropping date and lights columns
to_drop = df[['date','lights']]
df.drop(to_drop,axis=1,inplace=True)
df.head()
```

```python
# Appliances = Target variable!
```

```python
# check for correlation between variables::
df_corr = df.corr()
df_corr
```

```python
# correlation between the variables are not really very strong. So we keep the features as is
```

```python
# normalizing our dataset using MinMaxScaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
norm_arr = scaler.fit_transform(df)
columns = df.columns
norm_df = pd.DataFrame(norm_arr,columns=columns)
norm_df.head()
```

```python
# splitting the normalised dataset into feaure variables and target variable
target_col = norm_df['Appliances']
features_cols = norm_df.drop(['Appliances'],axis=1)
```

```python
# print five rows each from the target col and features cols::
print('Target col:')
target_col.head()
```

```python
print('Features col: ')
features_cols.head()
```

```python
# splitting our dataset into test set and training set for the regression algorithm

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    features_cols,target_col,test_size=0.3,random_state=42)
```

```python
# importing our linear regression algorithm
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

# fitting our model to out training set
linear_model.fit(X_train,y_train)

# obtaining prediction from our model; name 'y_pred'
y_pred = linear_model.predict(X_test)
```

```python
# Measuring regression performance.

# 1.Mean Absolute Error (MAE)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,y_pred)
print("Mean absolute error: {:.3f}".format(mae))
```

```python
# 2. R^2 error

from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
print("R squared error: {:.3f}".format(r2_score))
```

```python
# 3. Residual sum of squares

import numpy as np
rss = np.sum(np.square(y_test - y_pred))
print("Residual Sum of Squares: {:.3f}".format(rss))
```

```python
# 4. Mean squared error

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print("Mean squared error: {:.3f}".format(mse))

# 5. Root mean squared error
rmse = np.sqrt(mse)
print("Root mean squared error: {:.3f}".format(rmse))
```

```python
# 5. calculating score from off the model method:
pred_score = linear_model.score(X_test,y_test)
print("Model score: {:.3f}".format(pred_score))
```

### Regularising the linear model

```python
# >> Lasso regression

from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(X_train,y_train)

# >> Ridge regression

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(X_train,y_train)

```

```python
# comparing the effects of regularization:

def get_weights_df(model,features,col_name):
    # this function returns the weight of the features in the training dataset
    weights = pd.Series(linear_model.coef_,features_cols.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index() # converts the pandas series to pandas dataframe
    weights_df.columns = ['Features',col_name]
    weights_df[col_name].round(3)
    
    return weights_df

linear_model_weights = get_weights_df(linear_model,X_train,'Linear_model_weight')
ridge_weights_df = get_weights_df(ridge_reg,X_train,'Ridge_weight')
lasso_weights_df = get_weights_df(lasso_reg,X_train,'Lasso_weight')

final_weights_df = pd.merge(linear_model_weights,ridge_weights_df,on='Features')
final_weights_df = pd.merge(final_weights_df,lasso_weights_df,on='Features')
```
